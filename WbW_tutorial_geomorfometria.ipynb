## Trabalhando com CopDEM

O arquivo TIF está disponível em: https://github.com/CarlosGrohmann/whitebox_scripts

```python
cop_tif = 'CopDEM_clip_petar.tif'
cop = cop_tif[:-4] # nome do arquivo sem extensão '.tif'   
```

```python
# Reading raster data
dem = wbe.read_raster(cop_tif)
```

```python
dem
```

```python
dir(dem)
```

```python
dir(dem.configs)
```

```python
# https://www.whiteboxgeo.com/manual/wbw-user-manual/book/working-with-raster-data.html

def print_metadata(rast):
    # The RasterConfigs of a Raster object contains useful metadata about the Raster.
    print(f'Rows: {rast.configs.rows}')
    print(f'Columns: {rast.configs.columns}')
    print(f'Resolution (x direction): {rast.configs.resolution_x}')
    print(f'Resolution (y direction): {rast.configs.resolution_y}')
    print(f'North: {rast.configs.north}')
    print(f'South: {rast.configs.south}')
    print(f'East: {rast.configs.east}')
    print(f'West: {rast.configs.west}')
    print(f'Min value: {rast.configs.minimum}')
    print(f'Max value: {rast.configs.maximum}')
    print(f'EPSG code: {rast.configs.epsg_code}') # 0 if not set
    print(f'Nodata value: {rast.configs.nodata}')
    # What data type are stored in raster grid cells?
    # See the RasterDataType class for more info.
    print(f'Data type: {rast.configs.data_type}') 
    # What is the photometric interpretation, continuous, categorical, RGB, etc.?
    # See the PhotometricInterpretation class for more info.
    print(f'Photometric interpretation: {rast.configs.photometric_interp}')
```

```python
print_metadata(dem)
```

#### relevo sombreado

```python
# relevo sombreado
az = 45
alt = 35
zf = 1

shade = wbe.hillshade(dem=dem, azimuth=az, altitude=alt, z_factor=zf)
```

```python
shade
```

```python
shade_tif = f'{cop}_shade_{az}_{alt}_z{zf}.tif'
wbe.write_raster(shade, shade_tif, compress=True)

# opção - calcular e salvar tif em uma só linha
wbe.write_raster(wbe.hillshade(dem=dem, azimuth=az, altitude=alt, z_factor=zf), shade_tif, compress=True)
```

#### geomorfometria

```python
# declividade
slope_tif = f'{cop}_slope.tif'
slope = wbe.slope(dem=dem, units="percent")

wbe.write_raster(slope, slope_tif, compress=True)
```

```python
print_metadata(slope)
```

```python
# aspecto
aspect_tif = f'{cop}_aspect.tif'
aspect = wbe.aspect(dem=dem)

wbe.write_raster(aspect, aspect_tif, compress=True)
```

```python
# curvaturas
planc_tif = f'{cop}_plan_curvature_log.tif'
profc_tif = f'{cop}_profile_curvature_log.tif'
meanc_tif = f'{cop}_mean_curvature_log.tif'

planc = wbe.plan_curvature(dem=dem, log_transform=True)
profc = wbe.profile_curvature(dem=dem, log_transform=True)
meanc = wbe.mean_curvature(dem=dem, log_transform=True)

wbe.write_raster(planc, planc_tif, compress=True)
wbe.write_raster(profc, profc_tif, compress=True)
wbe.write_raster(meanc, meanc_tif, compress=True)
```

```python
# rugosidade
win = 11
rough_tif = f'{cop}_rough_{win}x{win}.tif'
rough = wbe.standard_deviation_of_slope(dem=dem, filter_size=win)

wbe.write_raster(rough, rough_tif, compress=True)
```

```python
# geomorphons
dist = 35
gmorph_tif = f'{cop}_geomorphons_d{dist}.tif'
gmorph = wbe.geomorphons(dem=dem, search_distance=dist, output_forms=True)

wbe.write_raster(gmorph, gmorph_tif, compress=True)
```

```python
# landforms
penn_tif = f'{cop}_pennock.tif'
penn = wbe.pennock_landform_classification(dem=dem, slope_threshold=3.0, prof_curv_threshold=0.1, plan_curv_threshold=0.0)

wbe.write_raster(penn[0], penn_tif, compress=True)
```
